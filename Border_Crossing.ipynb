```python
class CrossBorder():
    
    def __init__(self, index_for_date, index_for_value):
        self.index_for_date=index_for_date
        self.index_for_value=index_for_value
        
    def read_file(self, file_name):
        count = 0
        mainList = []
        with open(file_name) as fp: 
            Lines = fp.readlines() 
            for line in Lines: 
                count += 1
                mainList.append(line.strip().split(','))
        return mainList

    def spec_col(mainList, column):
        columnNumber = None
        columnRes = []
        for col in range(0, len(mainList[0])):
            if mainList[0][col] == column:
                columnNumber = col
        if columnNumber == None:
            print("column '{}' does not exist.".format(column))
        else:
            for i in range(1, len(mainList)-1):
                columnRes.append(mainList[i][columnNumber])
                #print(mainList[i][columnNumber])
            return columnRes

    def col_unique(self,mainList, column):
        columnNumber = None
        columnRes = set()
        for col in range(0, len(mainList[0])):
            if mainList[0][col] == column:
                columnNumber = col
        if columnNumber == None:
            print("column '{}' does not exist.".format(column))
        else:
            for i in range(1, len(mainList)-1):
                columnRes.add(mainList[i][columnNumber])
                #print(mainList[i][columnNumber])
            return columnRes


    def find_avg(self, mainList, column):
        ''' 
        Port Name, State, Port Code, Border, Date, Measure
        '''
        columnNumber = None
        sumCol = 0
        for col in range(0, len(mainList[0])):
            if mainList[0][col] == column:
                columnNumber = col
        if columnNumber == None:
            print("column '{}' does not exist.".format(column))
        else:
            try:
                for i in range(1, len(mainList)):
                    sumCol += int(mainList[i][columnNumber])
                print('Average for the column {} is {} .'.format(column, sumCol/len(mainList)))
            except:
                print('ERROR!')


    def find_sum_value(self, mainList, column, condition):
        columnNumber = None
        sumVal = 0
        myDic = {}                     
        months = ['months','January', 'February', 'March', 'April', 'May', 'June', 'July', 'August',
                  'September', 'October', 'November', 'December']

        # find the column
        for col in range(0, len(mainList[0])):
            if mainList[0][col] == column:
                columnNumber = col

        for i in range(1, len(mainList)):
            if mainList[i][columnNumber] == condition:
                key = mainList[i][self.index_for_date]
                # only we need month value
                key = key[0:2]
                if key not in myDic:
                    myDic[key] = int(mainList[i][self.index_for_value])
                else:
                    myDic[key] = myDic.get(key)
                    myDic[key] = myDic.get(key) + int(mainList[i][self.index_for_value])  
        #return myDic        
        for k,v in myDic.items():
            print(months[int(k)],':',v)
            sumVal += v
        print("Sum for the column: '{}' with condition '{}' is: {} ".format(column, condition, sumVal))     
```
